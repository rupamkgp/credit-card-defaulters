# 03. Model Training & Feature Importance
We extract feature importances using Random Forest and train our baseline models.

```python
from src.preprocessing import preprocess_pipeline
from src.train import train_baselines
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier

X_train, y_train, X_test, y_test, scaler, features = preprocess_pipeline('../data/raw/augmented_credit_card_clients.csv')
rf = RandomForestClassifier(random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)

importances = rf.feature_importances_
indices = importances.argsort()[::-1][:20]

plt.figure(figsize=(10, 8))
sns.barplot(x=[importances[i] for i in indices], y=[features[i] for i in indices], palette='viridis')
plt.title('Top 20 Feature Importances (Random Forest)')
plt.xlabel('Importance Score')
plt.savefig('../reports/images/feature_importance.png', dpi=300)
plt.show()
```
